## Step 4 Bar Charts

Make a bar plot with Seaborn to see the average size of Kiva awards given to projects by country.

We have set up a figure for you to plot on. The `f` variable gives us access to the figure and `ax` gives us access to the axes.

Use `sns.barplot()` with the following arguments:

- `x` set to `country`
- `y` set to `loan_amount`
- data set to `df`

````python
f, ax = plt.subplots(figsize=(15, 10))
sns.barplot(x="country", y = "loan_amount", data=df)
````

### Adding $
To make the visualization more clear you can use the following code to so that the `loan_amount` ticks on the y-axis begin with a `$`. 

```python
import matplotlib.ticker as mtick
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)

```

Run the code in the cell below to see the `$` in action.

````python
import matplotlib.ticker as mtick

f, ax = plt.subplots(figsize=(15, 10))
sns.barplot(x="country", y = "loan_amount", data=df)
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick) 
````

## Step 5 Bar Charts With `hue`

To visualize the loan amount by country, also aggregated by another factor, add the `hue` parameter to your `sns.barplot()` and set it so that the visualization includes the nested category of gender.

Hint:<font color=white> Set hue equal the column name "gender"  </font>


````python
f, ax = plt.subplots(figsize=(15, 10))
# starting code
# sns.barplot(x="country", y = "loan_amount", data=df)
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)

sns.barplot(x="country", y = "loan_amount", hue = "gender", data=df)
````

## Step 9 Violin Plots

You can use nearly the identical syntax to create violin plots as you have used for box plots. Take this line of code from above:

```python
sns.boxplot(x="activity", y = "loan_amount", data=kiva_df)
```
To visualize the distribution of the exact same data as a violin plot you could pass the same parameters to `sns.violinplot()` instead of `sns.boxplot()`.

Change the code in the cell below so that the data is plotted as a violin plot instead of a barplot.

````python
f, ax = plt.subplots(figsize=(16, 10))
# starting code
#sns.boxplot(x="activity", y = "loan_amount", data=kiva_df)
sns.violinplot(x="activity", y = "loan_amount", data=kiva_df)
````

### Create a violin plot that visualizes the distribution of loan amount by country.

Hint:<font color=white> 
 You can use this box plot code, but change it into a violin plot using the same parameters.
    
 sns.boxplot(x="country", y = "loan_amount", data=kiva_df)   
 </font>

````python
sns.set_palette("Paired")
f, ax = plt.subplots(figsize=(16, 10))
sns.violinplot(x="country", y = "loan_amount", data=kiva_df)
````

## Step 10 Split Violin Plots

Use the `hue` and `split` parameters with `sns.violinplot()` to visualize the distribution of loan amount by country, split by gender. 

Hint:<font color=white> 
 hue should be set to "gender" and split should equal True 
 </font>


````python
sns.set_palette("Spectral")
f, ax = plt.subplots(figsize=(18, 12))
sns.violinplot(x="country", y = "loan_amount", hue = "gender", split = True, data=kiva_df)
````
