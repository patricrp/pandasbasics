## Step 4 Barplot

Make a bar plot with Seaborn to see how each of the five countrires stack up in rural, urban, and total areas having clean drinking water.

We have set up a figure for you to plot on. Use `sns.barplot()` with the following arguments:

- `x` set to `country`
- `y` set to `value`
- hue set to `area`
- data set to `df`

```python
f, ax = plt.subplots(figsize=(25, 15))
sns.barplot(x="country", y = "value", hue = "area", data=df)
```

Set a different color palette using `sns.set_palette()`. You can use any of the Color Brewer built in 
{{how to surface color options here best}}


Plot a barplot but this time setting `sns.barplot()` equal to `ax` so you can access the axes properties using `ax.set()` on the next line of code.

Rename the y-axis label something more descriptive than `value`. For example: `(%) of population using clean water`
 
Hint:  <font color=white>Use ax.set(ylabel="descriptive label") </font>

```python
# set color palette
sns.set_palette("Set1")
f, ax = plt.subplots(figsize=(25, 15))

# barplot goes here
ax = sns.barplot(x="country", y = "value", hue = "area", data=df)
# set new ylabel here
ax.set(ylabel="(%) of population using clean water")
```

## Step 5 Boxplots

make a boxplot to see the distribution {{ talk about distrubution in boxplots}}

```python
f, ax = plt.subplots(figsize=(16, 10))
sns.boxplot(x="country", y = "value", hue = "area", data=df)
```

## Step 6 Update DataFrame

Let's take the United States out of the DataFrame and make another boxplot so we can focus on the countries with a larger distribution. (and americans be grateful for such great access to clean drinking water!)

Remove any row that has a value of USA for country:
filter the data frame and assign back to the original data frame so that `df`is a copy of itself without the rows where the value for  `country` is `United States of America`

```python
# removing USA from DataFrame and trying again
df= df[df["country"] != "United States of America"]
df.head()
```

Rebox plot {{needs to be turned into instruction}}

```python
f, ax = plt.subplots(figsize=(16, 10))
sns.set_context("poster")
sns.set_palette("Paired")
sns.boxplot(x="country", y = "value", hue = "area", data=df)
```

## Step 7 A Different Boxplot View Of The Data 

Create a boxplot that shows the distribution of each of the three areas by country, with `area` on the x-axis, `value` on the right, and `hue` set to `country`.

```python
f, ax = plt.subplots(figsize=(16, 10))
sns.set_context("poster")
sns.boxplot(x="area", y = "value", hue = "country", data=df, palette="Set2")
```

How does this plot visualize the same data differently than the one above?

## Step 8 Violinplots

```python
# violin plot
f, ax = plt.subplots(figsize=(16, 10))
sns.set_context("poster")
sns.violinplot(x="area", y = "value", hue = "country", data=df, palette="Set2")
```

## Step 9 Focus In On Mexico

```python
# make a violin plot of just mexico
df_mexico= df[df["country"] == "Mexico"]
# df_mexico = df_mexico[df_mexico["area"] !="Total"]
df_mexico.head()
f, ax = plt.subplots(figsize=(16, 10))
sns.set_context("poster")
sns.violinplot(x="country", y = "value", hue = "area", data=df_mexico, palette="Set2")
```

```python
# make a violin plot of just mexico
# df_mexico= df[df["country"] == "Mexico"]
df_mexico = df_mexico[df_mexico["area"] !="Total"]
# df_mexico.head()
f, ax = plt.subplots(figsize=(16, 10))
sns.set_context("poster")
sns.violinplot(x="country", y = "value", hue = "area", data=df_mexico, split = True, palette="Set2")
```
